```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold

X = pd.read_csv('data250.csv', index_col='index')
```

```python
X
```

```python
variances = X.var()
variances.to_csv('var.csv', header=['Variance'])
variance_threshold = 0.1
selector = VarianceThreshold(threshold=variance_threshold)

X_high_variance = selector.fit_transform(X)
retained_features = X.columns[selector.get_support()]
new_data = pd.DataFrame(X_high_variance, index=X.index, columns=retained_features)
new_data.to_csv('cal_rd_data_high_variance.csv')
```
