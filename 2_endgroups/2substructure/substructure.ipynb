```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('plot.csv')
bar_width = 0.35
index = range(len(data))

bars1 = plt.bar(index, data['250 data'], width=bar_width, label='250 data', color='#EE8027', align='center')
bars2 = plt.bar([i + bar_width for i in index], data['1000 data'], width=bar_width, label='1000 data', color='#4078B1', align='center')

for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), ha='center', va='bottom', fontsize=8.5)

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), ha='center', va='bottom', fontsize=8.5)

plt.xlabel('Substructures', fontsize=14, fontweight='bold')
plt.ylabel('Counts', fontsize=14, fontweight='bold')
plt.xticks([i + bar_width / 2 for i in index], data['Substructures'], rotation=45)

plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['right'].set_linewidth(2)
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['left'].set_linewidth(2)
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.legend()
plt.ylim(0, 650)
plt.tight_layout()
plt.savefig('substructure_count.jpg', dpi=1000, bbox_inches='tight', pad_inches=0.1)
plt.show()
```
